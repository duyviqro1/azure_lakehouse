### Nhập dữ liệu các file CSV vào Delta table

```python
files_name = ['netflix_titles', 'netflix_directors', 'netflix_countries', 'netflix_category', 'netflix_cast']
```

```python
def autoloader(file_name):
    checkpoint_location = f"abfss://checkpoint@netflixprojectdltp.dfs.core.windows.net/{file_name}"

    df = spark.readStream\
        .format("cloudFiles")\
        .option("cloudFiles.format", "csv")\
        .option("cloudFiles.schemaLocation", checkpoint_location)\
        .load("abfss://raw@netflixprojectdltp.dfs.core.windows.net/" + file_name)

    df.writeStream\
        .option("checkpointLocation", checkpoint_location)\
        .trigger(availableNow=True)\
        .toTable(f'netflix_catalog.bronze.{file_name}')
```

```python
for file_name in files_name:
    autoloader(file_name)
```
